# 初始量化交易

量化交易，量化交易是指借助现代统计学和数学的方法，利用计算机技术来进行交易的证券投资方式。

- 量化交易相关知识
    + 股票
    + 金融、数学、计算机
    
<br>
    
- 量化交易怎么做
    + 交易策略
    + 把策略转化为程序
    + 验证策略
        * 回测
        * 模拟交易
        
回测： 让计算机能根据一段时间的历史数据模拟执行该策略，根据结果评价并改进策略。

模拟交易： 让计算机能根据实际行情模拟执行该策略一段时间，根据结果评价并改进策略。

与回测不同，回测是用历史数据模拟，模拟交易使用实际的实时行情来模拟执行策略的。

<br>

- 量化交易需要什么
    + 各种数据
    + 量化交易系统

 **K线(Candelstick chart)**： 
 又称阴阳烛，是反映价格走势的一种图线。其特色在于一个线段内记录了多项信息，相当易读且实用有效。广泛用于股票、期货、贵金属、数字货币等行情的技术分析，称为K线分析。
 
 - 阳线： 收盘价大于开盘价，以红色表示
 - 阴线： 收盘价低于开盘价，以绿色表示
 - 中立线： 收盘价等于开盘价

**移动平均线(moving average, MA)**：
简称均线，是技术分析中一种分析时间序列数据的工具。最常见的是利用股价、回报或交易量等变数计算出移动平均线。

**市价单(Market Order)**：
以当时市场价格成交的订单。

**限价单(Limit Order)**：
指买入或卖出必须在一个指定的价格成交。

**止损单(Stop Order)**：
指在订单中设置止损价格，一旦股价到达所设定的止损价，将会以市价单的方式成交。
